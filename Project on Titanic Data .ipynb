# Level 3 – Grouping & Aggregation

```python
#Calculate the average age by gender.
# Average age by gender
df.groupby('Sex')['Age'].mean()
```

```python
#Find the survival rate by passenger class.
survival_rate = df.groupby('Pclass', as_index=False)['Survived'].mean()
survival_rate['Survived'] = (survival_rate['Survived'] * 100).round(2)
survival_rate.rename(columns={'Survived': 'Survival_Rate (%)'}, inplace=True)
print(survival_rate)

```

```python
#Get the total number of siblings/spouses by embarkation point.
sibsp_sum = df.groupby('Embarked', as_index=False)['SibSp'].sum()
sibsp_sum.rename(columns={'SibSp': 'Total_Siblings_Spouses'}, inplace=True)
print(sibsp_sum)
```

```python
#Find the median fare for each passenger class.
median_fare = df.groupby('Pclass', as_index=False)['Fare'].median()
median_fare.rename(columns={'Fare': 'Median_Fare'}, inplace=True)
print(median_fare)
```

```python
#Find the maximum age for each Embarked category.
max_age = df.groupby('Embarked', as_index=False)['Age'].max()
max_age.rename(columns={'Age': 'Max_Age'}, inplace=True)
print(max_age)
```

# Level 4 – Data Cleaning

```python
#Fill missing Age values with median age
df['Age'] = df['Age'].fillna(df['Age'].median())
print(df['Age'])
```

```python
# Replace missing Embarked values with the most frequent (mode) value
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
print(df["Embarked"])
```

```python
# Create a new column FamilySize = SibSp + Parch + 1
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print(df['FamilySize'])
```

```python
#Extract title (Mr., Mrs., Miss, etc.) from Name
df['Title'] = df['Name'].str.extract(r',\s*([^\.]*)\.', expand=False).str.strip()
print(df['Title'])
```

```python
import matplotlib.pyplot as plt

# 1. Bar chart: passenger count by Pclass
df['Pclass'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

```

```python
# 2. Histogram: passenger ages
df['Age'].plot(kind='hist', bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
